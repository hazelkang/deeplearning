# 2. 객체 지향 방식

## 1) plt.subplots

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html?highlight=subplots#matplotlib.pyplot.subplots

### 0> shift+tab

- nrows와 ncols로 subplot 배치할 행과 열을 생성

- return을 fig와 ax로 2개한다 (fix와 ax라는 객체를 만들어서 진행 => 객체 지향)

- fig => figure

- ax => axes

```python
fig, ax = plt.subplots(1,2)
```

### 1> 각 그래프 설정 (ax)

- nd array indexing 하듯 각 그래프를 지칭하면 됩니다.

- set을 앞에 붙여서 옵션을 추가합니다.

```python
fig, ax = plt.subplots(2,2)
ax[0,0].set_title('title 1')
ax[0,0].plot([1,2,3])
ax[0,1].set_title('title 2')
```

### 2> 이미지 사이즈 조절

- figsize로 figure 크기 확보

```python
fig, ax = plt.subplots(2,2, figsize = (10,10))
ax[0,0].set_title('title 1')
ax[0,1].set_title('title 2')
```
