## 1) plt.add_subplot

https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html?highlight=add_subplot#matplotlib.figure.Figure.add_subplot

### 0> shift+tab

- nrows와 ncols로 subplot 배치할 행과 열을 생성하고 <font color="red"> index까지 설정 (쉼표를 붙이기도 붙이지 않기도 한다.)

```python
fig = plt.figure() # Figure 객체 생성
ax = fig.add_subplot(211) # 2행 1열에서 1번째 index에 배치한다.
ax.plot('title')
```

여기까지만 봐서는 객체 지향인 것 같을 수 있습니다.

### 1> state-machine과 객체지향이 혼용되어 사용

state-machine이라면 한 ax에 겹쳐서 그려져야 하고

객체지향이면 하나만 실행되어야 하는데 

이들의 중간정도로 실행됩니다.



```python
fig = plt.figure()
ax = fig.add_subplot(2,1,1)
ax.plot([1,2,3],'b')

ax = fig.add_subplot(2,1,2)
ax.plot([3,4,5],'r')
```

```python
fig = plt.figure()
ax = fig.add_subplot(2,1,1)
```

### 1> 각 그래프 설정 (ax)

- nd array indexing 하듯 각 그래프를 지칭하면 됩니다.

- set을 앞에 붙여서 옵션을 추가합니다.

```python
fig, ax = plt.subplots(2,2)
ax[0,0].set_title('title 1')
ax[0,1].set_title('title 2')
```

### 2> 여유공간 확보

- figsize로 figure 크기 확보

- tight_layout => 여백을 조절한다.

```python
fig, ax = plt.subplots(2,2, figsize = (10,10))
ax[0,0].set_title('title 1')
ax[0,1].set_title('title 2')
plt.tight_layout(pad=5)
```
